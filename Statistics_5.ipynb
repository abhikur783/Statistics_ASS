```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Ans:
    
Estimation statistics is a branch of inferential statistics that involves making educated guesses or estimates about population
parameters based on sample data. When it is not feasible or practical to measure an entire population, statisticians use sampl-
ing techniques to collect data from a subset of the population. Estimation allows us to draw conclusions about the entire popul-
ation using this sample data.

There are two main types of estimates in estimation statistics: point estimate and interval estimate.

Point Estimate:
A point estimate is a single value that serves as the best guess for the unknown population parameter. It is obtained by using a
statistic calculated from the sample data that is assumed to be a reasonable representation of the population parameter. Common
point estimates include the sample mean, sample proportion, sample median, and sample variance, among others. For example, if 
you want to estimate the average height of all students at a school, you can calculate the sample mean height of a random sample
of students and use it as the point estimate for the population mean height.
However, it's important to note that point estimates do not give any information about the precision or uncertainty associated 
with the estimate. This is where interval estimates come into play.

Interval Estimate:
An interval estimate is a range of values within which the population parameter is expected to lie, along with a degree of conf-
idence in the estimate. It provides a measure of uncertainty associated with the point estimate. The interval estimate is typic-
ally expressed as a confidence interval. A confidence interval consists of two numbers: a lower bound and an upper bound.
For example, if a 95% confidence interval for the average height of students at a school is 160 cm to 165 cm, it means that we 
are 95% confident that the true population mean height lies within this range. The wider the confidence interval, the less cert-
ain we are about the estimate; whereas, a narrower interval indicates a more precise estimate.

Interval estimates take into account both the variability in the sample data and the sample size. As the sample size increases, 
the interval tends to become narrower, indicating increased precision in the estimate.

In summary, estimation statistics involves using sample data to make educated guesses about population parameters. Point estima-
tes provide a single value as the best guess for the parameter, while interval estimates provide a range of values along with a
degree of confidence in the estimate, reflecting the uncertainty associated with the point estimate.
```

```python
Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
    Ans:
        Hypothesis testing is a fundamental concept in statistics used to make decisions about the characteristics of a popul-
        ation based on sample data. It involves testing a claim or hypothesis about a population parameter, such as the popula-
        tion mean, population proportion, population standard deviation, etc.

The process of hypothesis testing generally involves the following steps:

Formulating the Hypotheses:

Null Hypothesis (H0): This is the default assumption or claim. It represents the status quo or no effect and is often denoted as
    "there is no difference" or "there is no effect."
Alternative Hypothesis (Ha): This is the claim that contradicts the null hypothesis. It represents the effect or change we are 
    interested in investigating.
Collecting Sample Data:

Data is collected from a random sample from the population of interest.
Calculating Test Statistic:

A test statistic is calculated using the sample data and a specific statistical formula based on the hypothesis being tested.

Determining the Rejection Region or P-value:

Based on the null hypothesis, a critical region (rejection region) is defined. If the test statistic falls within this region, 
the null hypothesis is rejected.
Alternatively, the p-value is calculated, which represents the probability of observing the sample data, assuming the null hypo-
thesis is true. If the p-value is below a pre-defined significance level (commonly denoted as alpha), the null hypothesis is 
rejected.
Making a Decision:

Based on the test statistic or p-value, a decision is made regarding whether to reject the null hypothesis or not.
If the null hypothesis is rejected, it means there is evidence to support the alternative hypothesis. If the null hypothesis is
not rejected, there is not enough evidence to support the alternative hypothesis.
Importance of Hypothesis Testing:

Inference about Populations: Hypothesis testing allows us to draw conclusions about population parameters based on sample data.
    It helps researchers and decision-makers make informed judgments about population characteristics.

Scientific Research: Hypothesis testing is a crucial tool in scientific research. It enables researchers to test specific hypo-
    theses and theories, allowing for the advancement of knowledge and understanding in various fields.

Data-Driven Decision Making: Hypothesis testing provides a systematic and objective way to make decisions based on data. It hel-
    ps in making evidence-based choices in areas like business, healthcare, social sciences, and more.

Identifying Significant Relationships: Hypothesis testing helps identify significant relationships between variables. For examp-
    le, in medical studies, it helps determine whether a new treatment is effective compared to a placebo.

Assessing Real-World Scenarios: Hypothesis testing is used in quality control and manufacturing to assess whether a product mee-
    ts certain specifications or standards.
```

```python
Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.
Ans:
    import numpy as np
from scipy.stats import ttest_ind

def hypothesis_test_two_means(sample1, sample2, alpha=0.05, alternative='two-sided'):
    """
    Perform a two-sample t-test for the difference between two population means.

    Parameters:
        sample1 (array-like): The sample data from population 1.
        sample2 (array-like): The sample data from population 2.
        alpha (float): The significance level (default is 0.05).
        alternative (str): The alternative hypothesis ('two-sided', 'greater', or 'less').

    Returns:
        t_statistic (float): The calculated t-statistic.
        p_value (float): The two-tailed p-value.
    """
    t_statistic, p_value = ttest_ind(sample1, sample2, alternative=alternative, equal_var=False)
    
    return t_statistic, p_value

# Example usage:
sample1 = [65, 68, 72, 74, 70]  # Sample data from population 1 (e.g., male college students' weights)
sample2 = [60, 62, 65, 64, 63]  # Sample data from population 2 (e.g., female college students' weights)

t_statistic, p_value = hypothesis_test_two_means(sample1, sample2)

print("Results of Hypothesis Test:")
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The average weight of male college students is greater than the average weight of female 
          college students.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average weight between male and female col-lege students.")
```

```python
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.
Ans:
    import numpy as np
from scipy.stats import t

def two_sample_t_test(sample1, sample2):
    """
    Calculate the two-sample t-test for independent samples assuming equal variances.

    Parameters:
        sample1 (array-like): The first sample data.
        sample2 (array-like): The second sample data.

    Returns:
        t_statistic (float): The calculated t-statistic.
        p_value (float): The two-tailed p-value.
    """
    n1 = len(sample1)
    n2 = len(sample2)
    dof = n1 + n2 - 2  # Degrees of freedom (pooled variance)

    # Calculate sample means and variances
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate the pooled standard deviation
    pooled_std_dev = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / dof)

    # Calculate the t-statistic
    t_statistic = (mean1 - mean2) / (pooled_std_dev * np.sqrt(1 / n1 + 1 / n2))

    # Calculate the two-tailed p-value
    p_value = 2 * (1 - t.cdf(abs(t_statistic), dof))

    return t_statistic, p_value

# Example usage:
sample1 = [23, 34, 28, 35, 29]
sample2 = [31, 27, 30, 33, 25]

t_statistic, p_value = two_sample_t_test(sample1, sample2)

print("Results of Two-Sample t-test:")
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The population means are not equal.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in population means.")
```

```python
Q12: What is t-statistic? State the formula for t-statistic.
    Ans:
The t-statistic is a measure used in hypothesis testing to determine the difference between sample means and assess whether that
difference is statistically significant.

Formula for t-statistic (for two independent samples assuming equal variances):
\[ t = \frac{{\text{mean of sample 1} - \text{mean of sample 2}}}{{\text{pooled standard deviation} \times \sqrt{\frac{1}{n_1} +
                                                                                                            \frac{1}{n_2}}}} \]

where:
- mean of sample 1: The mean of the first sample.
- mean of sample 2: The mean of the second sample.
- pooled standard deviation: A combined estimate of the standard deviation, assuming equal variances in both samples.
- \( n_1 \): The size of the first sample.
- \( n_2 \): The size of the second sample.

The t-statistic measures how many standard errors the sample means differ from each other. It is used to calculate the p-value, 
which helps determine whether the observed difference between sample means is statistically significant or merely due to random
chance.
```
