```python
Q1. What is the Probability density function?
Ans:
    
The Probability Density Function (PDF) is a fundamental concept in probability theory and statistics. It is used to describe the
likelihood of a continuous random variable taking on a particular value or falling within a specific range of values. In simpler
terms, the PDF gives us the probability distribution of a continuous random variable.

For a continuous random variable X, the PDF is denoted as f(x) and is defined such that the probability of X falling within a 
certain interval [a, b] is given by the integral of the PDF over that interval:

P(a ≤ X ≤ b) = ∫(from a to b) f(x) dx

The properties of the PDF are as follows:

f(x) is non-negative: f(x) ≥ 0 for all x in the domain of X.
The total area under the PDF curve is equal to 1: ∫(over the entire range) f(x) dx = 1.
Probability of X falling in a particular interval: The probability of X lying within a specific interval [a, b] can be found by 
    integrating the PDF over that interval.
It's important to note that the PDF is only applicable to continuous random variables. For discrete random variables, we use a 
Probability Mass Function (PMF) instead, which gives the probabilities of individual discrete values.
```

```python
Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.
Ans:
    
write a Python function to calculate the probability density function (PDF) of a normal distribution at a given point. The PDF 
of a normal distribution is given by the formula:

f(x) = (1 / (σ * √(2 * π))) * exp(-(x - μ)^2 / (2 * σ^2))

where:

μ is the mean of the normal distribution.
σ is the standard deviation of the normal distribution.
π is the mathematical constant Pi.
exp() is the exponential function.
To calculate the PDF at a given point x, you can use this formula in the following Python function:

python code
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
        x (float): The point at which to calculate the PDF.
        mean (float): Mean of the normal distribution.
        std_dev (float): Standard deviation of the normal distribution.

    Returns:
        float: The value of the PDF at the given point 'x'.
    """
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

# Example usage:
mean = 0
std_dev = 1
point = 1.5
pdf_at_point = normal_pdf(point, mean, std_dev)
print("PDF at x = {}: {}".format(point, pdf_at_point))
Replace mean, std_dev, and point with the desired values for your specific normal distribution and point of interest. The 
function will return the probability density function value at that given point.
```

```python
Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.
Ans:
The Binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of inde-
pendent Bernoulli trials. The properties of the Binomial distribution are as follows:

Fixed Number of Trials: The Binomial distribution is used when there is a fixed number of independent trials, denoted by 'n'.

Independent Trials: Each trial must be independent of the others, meaning the outcome of one trial does not affect the outcome 
    of another.

Two Possible Outcomes: Each trial has two possible outcomes, often referred to as "success" and "failure."

Constant Probability: The probability of success (denoted by 'p') remains constant across all trials.

Discrete Values: The Binomial random variable counts the number of successes, which must be a whole number between 0 and 'n'.

The probability mass function (PMF) of the Binomial distribution is given by:

P(X = k) = C(n, k) * p^k * (1 - p)^(n - k)

where:

P(X = k) is the probability of getting exactly 'k' successes in 'n' trials.
C(n, k) is the binomial coefficient, also denoted as "n choose k," which is the number of ways to choose 'k' successes out of
'n' trials.
p is the probability of success in a single trial.
(1 - p) is the probability of failure in a single trial.
Two examples of events where the Binomial distribution can be applied are:

Coin Tossing: When flipping a fair coin (with equal probabilities of getting heads or tails) 'n' times, we can use the Binomial 
    distribution to calculate the probabilities of getting a specific number of heads in those 'n' tosses.

Quality Control: In manufacturing processes, we might be interested in the number of defective items in a batch of 'n' items. If
    the probability of an item being defective remains constant throughout the production, we can use the Binomial distribution 
    to calculate the probabilities of getting a certain number of defective items in the batch.
```

```python
"""Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.
Ans:
To generate a random sample of size 1000 from a binomial distribution with a probability of success (p) of 0.4 and plot a 
histogram of the results using matplotlib, you can use the numpy library to generate the random sample and the matplotlib.pyplot
library to create the histogram. Make sure you have both libraries installed by running pip install numpy matplotlib if you 
haven't installed them already. Here's a Python code to accomplish this task:

python code"""
import numpy as np
import matplotlib.pyplot as plt

def generate_binomial_sample(size, p):
    """
    Generate a random sample from a binomial distribution.

    Parameters:
        size (int): Size of the random sample.
        p (float): Probability of success in a single trial.

    Returns:
        numpy array: The generated random sample.
    """
    return np.random.binomial(n=1, p=p, size=size)

# Generate the random sample
sample_size = 1000
probability_of_success = 0.4
random_sample = generate_binomial_sample(sample_size, probability_of_success)

# Create the histogram
plt.hist(random_sample, bins=2, alpha=0.6, color='g', edgecolor='black')
plt.xlabel("Success (1) or Failure (0)")
plt.ylabel("Frequency")
plt.title("Histogram of Random Sample from Binomial Distribution")
plt.xticks([0, 1], ['Failure', 'Success'])
plt.grid(axis='y')
plt.show()
```

```python
Q7. How Binomial distribution different from Poisson distribution?
Ans:
The Binomial distribution and the Poisson distribution are both discrete probability distributions, but they are used to model 
different types of events and have distinct characteristics. Here are the main differences between the two:

Nature of Events:

Binomial Distribution: The Binomial distribution is used to model the number of successes in a fixed number of independent 
    Bernoulli trials. Each trial has two possible outcomes (usually labeled as success and failure).
Poisson Distribution: The Poisson distribution is used to model the number of events occurring in a fixed interval of time or 
    space when the events happen at a constant rate and independently of the time since the last event.
Number of Trials/Events:

Binomial Distribution: Requires a fixed number of independent trials (n) to be specified in advance.
Poisson Distribution: Does not require a fixed number of trials or events. It models the number of events in a continuous or 
    infinite space or time.
Parameters:

Binomial Distribution: The Binomial distribution is defined by two parameters: the probability of success in a single trial (p)
        and the number of trials (n).
Poisson Distribution: The Poisson distribution is defined by a single parameter: the average rate of events per interval (λ).
Range of Values:

Binomial Distribution: The random variable in the Binomial distribution represents the number of successes and can take on 
    integer values between 0 and n.
Poisson Distribution: The random variable in the Poisson distribution represents the number of events and can take on non-
    negative integer values (0, 1, 2, 3, ...).
Assumptions:

Binomial Distribution: Assumes that each trial is independent of the others and that the probability of success (p) remains 
    constant across all trials.
Poisson Distribution: Assumes that events occur randomly and independently, and the average rate of occurrence (λ) is constant
    over time or space.
Shape:

Binomial Distribution: The shape of the Binomial distribution depends on the number of trials (n) and the probability of success
    (p).
Poisson Distribution: The shape of the Poisson distribution depends on the average rate of occurrence (λ), and it tends to have
    a peak around the mean λ.
In summary, the Binomial distribution is used to model the number of successes in a fixed number of trials with two possible 
outcomes, while the Poisson distribution is used to model the number of events occurring in a continuous or infinite space or
time, assuming a constant rate of occurrence.
```

```python
"""Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.
Ans:
To generate a random sample of size 1000 from a Poisson distribution with a mean of 5 and calculate the sample mean and variance
, you can use the numpy library to generate the random sample and then compute the mean and variance using NumPy functions.
Here's how you can do it in Python:

python code"""
import numpy as np

def generate_poisson_sample(size, mean):
    """
    Generate a random sample from a Poisson distribution.

    Parameters:
        size (int): Size of the random sample.
        mean (float): Mean of the Poisson distribution.

    Returns:
        numpy array: The generated random sample.
    """
    return np.random.poisson(lam=mean, size=size)

# Generate the random sample
sample_size = 1000
mean = 5
random_sample = generate_poisson_sample(sample_size, mean)

# Calculate the sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample, ddof=1)  # ddof=1 to compute the sample variance (N-1)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

```python
Q9. How mean and variance are related in Binomial distribution and Poisson distribution?
Ans:
In both the Binomial distribution and the Poisson distribution, the mean and variance are related, but the nature of this 
relationship is different for each distribution.

Binomial Distribution:
In a Binomial distribution with parameters n (number of trials) and p (probability of success in a single trial), the mean (μ) 
and variance (σ^2) are related as follows:

Mean (μ) = n * p
Variance (σ^2) = n * p * (1 - p)

In other words, the mean of a Binomial distribution is equal to the product of the number of trials (n) and the probability of
success in a single trial (p). The variance is the product of n, p, and (1 - p).

Poisson Distribution:
In a Poisson distribution with parameter λ (average rate of events per interval), the mean (μ) and variance (σ^2) are related as
follows:

Mean (μ) = λ
Variance (σ^2) = λ

In the Poisson distribution, both the mean and the variance are equal to the parameter λ.

Comparison:

In the Binomial distribution, the variance depends on both the number of trials (n) and the probability of success (p). As the 
probability of success (p) moves away from 0.5 (toward either 0 or 1), the variance tends to decrease, reaching its minimum when
p = 0 or p = 1 (i.e., when the distribution becomes degenerate).
In the Poisson distribution, the variance is equal to the mean (λ). Unlike the Binomial distribution, there is no second
parameter affecting the variance.
In summary, the mean and variance of both distributions are directly related, but the specific relationship differs for the
Binomial distribution and the Poisson distribution. The variance of the Binomial distribution depends on both the number of 
trials and the probability of success, while the variance of the Poisson distribution is solely determined by the average rate
of events per interval.
```

```python
Q10. In normal distribution with respect to mean position, where does the least frequent data appear?
Ans:
    
In a normal distribution, the least frequent data appears in the tails of the distribution, which are the regions further away
from the mean. The normal distribution is a symmetric bell-shaped curve, and its shape is characterized by the mean (μ) and 
standard deviation (σ).

Here's how the data distribution is in a normal distribution:

Mean (μ): The mean of a normal distribution represents the center of the distribution. It is the point around which the data is 
    symmetrically distributed. The mean is also the point of highest frequency, meaning the most common value in the distribut-
    ion.

Tails: As you move away from the mean in both directions, towards positive and negative extremes, the frequency of data decreas-
    es. The tails of the normal distribution represent the least frequent data points.

Standard Deviation (σ): The standard deviation controls the spread of the distribution. A larger standard deviation means the data points are more spread out, and the tails are wider. A smaller standard deviation means the data points are closer together, and the tails are narrower.

In a standard normal distribution (mean = 0, standard deviation = 1), about 68% of the data falls within one standard deviation of the mean (between -1 and +1), about 95% falls within two standard deviations (between -2 and +2), and about 99.7% falls within three standard deviations (between -3 and +3).

The data in the tails (beyond three standard deviations from the mean) is considered to be extreme and occurs less frequently, which is why it is referred to as the least frequent data in the normal distribution.
```
