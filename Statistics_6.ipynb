```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test. 
Ans:
    
A1: The main difference between a t-test and a z-test lies in the population parameters and sample size used for the inference.
    

T-test:
Used when the population standard deviation is unknown or when the sample size is small (typically n < 30).
Example Scenario: Suppose you want to compare the average scores of two groups of students (Group A and Group B) on a math test.
    You have collected sample data for both groups, and the sample size is relatively small (e.g., 20 students in each group). 
    In this case, you would use a t-test to determine if there is a significant difference in the mean scores between the two 
    groups.
Z-test:
Used when the population standard deviation is known or when the sample size is large (typically n ≥ 30).
Example Scenario: Let's say you have access to the average height of all adult males in a particular country, and you want to 
    compare the average height of a sample of 100 adult males from a different city with this known population average. Since 
    you know the population standard deviation and the sample size is large, you can use a z-test to determine if the sample's 
    average height significantly differs from the known population average.
In summary, t-tests are appropriate for small sample sizes or when the population standard deviation is unknown, while z-tests 
are suitable for large sample sizes when the population standard deviation is known.
```

```python
Q2: Differentiate between one-tailed and two-tailed tests. 
Ans:

One-tailed test: In a one-tailed test, the hypothesis is directional, and we are interested in whether the sample statistic is 
    significantly greater than or significantly less than a certain value. The critical region is on only one side of the distr-
    ibution.

Example: Testing if a new drug improves test scores, with the one-tailed hypothesis being that the drug will increase test scor-
    es.

Two-tailed test: In a two-tailed test, the hypothesis is non-directional, and we are interested in whether the sample statistic
    is significantly different from a certain value, regardless of the direction. The critical region is split into two sides of
    the distribution.

Example: Testing if a coin is fair, with the two-tailed hypothesis being that the coin is not biased and has an equal chance of 
    landing heads or tails.

In summary, one-tailed tests are used when we have a specific directional expectation, whereas two-tailed tests are used when
we are simply interested in detecting any significant difference, regardless of the direction.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error. 
Ans:

Type 1 error (False Positive): This occurs when we incorrectly reject a true null hypothesis. In other words, we mistakenly con-
    clude that there is a significant effect or difference when there is actually none in the population.
Example: In a clinical trial, a pharmaceutical company is testing a new drug to determine if it's more effective than a placebo
    in treating a certain condition. A Type 1 error would happen if the study incorrectly concludes that the drug is effective 
    (rejects the null hypothesis) when, in reality, the drug has no therapeutic effect.

Type 2 error (False Negative): This occurs when we fail to reject a false null hypothesis. In other words, we miss detecting a
    significant effect or difference that actually exists in the population.
Example: In airport security, metal detectors are used to identify passengers carrying metal objects that may be potential weap-
    ons. A Type 2 error would happen if the metal detector fails to identify a person carrying a concealed weapon (fails to rej-
    ect the null hypothesis that the person is not carrying a weapon) when, in reality, the person is carrying a dangerous item.
    

In summary, Type 1 error involves mistakenly detecting an effect that doesn't exist, while Type 2 error involves missing a real
effect or difference. Both errors are important considerations in hypothesis testing and can have significant implications depe-
nding on the context of the study or test.
```

```python
Q4: Explain Bayes's theorem with an example.
    Ans:
        Bayes's Theorem is a mathematical formula that allows us to update the probability of an event based on new evidence. 
        It relates the probability of an event A given the occurrence of event B to the probability of event B given the occu-
        rrence of event A.

The formula for Bayes's Theorem is:

\[ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} \]

where:
- \( P(A|B) \) is the probability of event A occurring given that event B has occurred.
- \( P(B|A) \) is the probability of event B occurring given that event A has occurred.
- \( P(A) \) is the prior probability of event A (the initial probability of A occurring).
- \( P(B) \) is the prior probability of event B (the initial probability of B occurring).

Example:
Suppose there is a rare disease that affects 1 in 1000 people in a certain population. A medical test has been developed to det-
ect this disease, and it is 95% accurate, meaning it correctly identifies a person with the disease 95% of the time and gives a 
false negative (saying a person without the disease has it) 5% of the time.

Let's use Bayes's Theorem to calculate the probability that a person has the disease given that the test result is positive (B) 
and the prior probability of having the disease (A) is 1 in 1000 (0.001).

- \( P(B|A) \) (the probability of testing positive given that a person has the disease) is 0.95.
- \( P(A) \) (the prior probability of having the disease) is 0.001.
- \( P(B) \) (the probability of testing positive) can be calculated using the law of total probability:
  \( P(B) = P(B|A) \times P(A) + P(B|\neg A) \times P(\neg A) \)
  where \( P(B|\neg A) \) is the probability of testing positive given that a person does not have the disease (1 - 0.95 = 0.05)
    , and \( P(\neg A) \) is the probability of not having the disease (1 - 0.001 = 0.999).
  \( P(B) = 0.95 \times 0.001 + 0.05 \times 0.999 \)
  \( P(B) \approx 0.0509 \)

Now, we can apply Bayes's Theorem:

\[ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} \]
\[ P(A|B) = \frac{0.95 \times 0.001}{0.0509} \]
\[ P(A|B) \approx 0.0187 \]

So, the probability that a person has the disease given a positive test result is approximately 0.0187 or 1.87%.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
    Ans:
        A confidence interval is a statistical range that provides an estimate of the uncertainty around a population parameter,
        such as the mean or proportion, based on a sample of data. It indicates a range of values within which we can reasonably
        expect the true population parameter to lie with a certain level of confidence.

To calculate a confidence interval, you need three main components:
1. Sample data: A representative sample from the population of interest.
2. Confidence level: The desired level of confidence, often expressed as a percentage (e.g., 95% confidence level).
3. Statistical method: The appropriate statistical formula based on the sample size and the population parameter being estimated
    .

Example:
Let's say we want to estimate the average height of adult males in a certain town. We collect a random sample of 100 adult males
and measure their heights. The sample mean height is 175 cm, and the sample standard deviation is 5 cm.

To calculate a 95% confidence interval for the population mean height, we use the formula for the confidence interval of the me-
an:

\[ \text{Confidence Interval} = \text{Sample Mean} \pm (\text{Critical Value} \times \frac{\text{Sample Standard Deviation}}{\s-
                                qrt{\text{Sample Size}}}) \]

First, we need to find the critical value corresponding to a 95% confidence level. For a large sample size (typically above 30)
and a normal distribution, the critical value is approximately 1.96.

Now, we can calculate the confidence interval:

\[ \text{Confidence Interval} = 175 \pm (1.96 \times \frac{5}{\sqrt{100}}) \]
\[ \text{Confidence Interval} = 175 \pm 0.98 \]
\[ \text{Confidence Interval} = (174.02, 175.98) \]

Thus, with 95% confidence, we can say that the true average height of adult males in the town lies between 174.02 cm and 175.98 
cm.
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.
Ans:
    Sample Problem:
A factory produces two types of light bulbs: Type A and Type B. The quality control data shows that 10% of the bulbs produced 
    are Type A, while 90% are Type B. The factory's past records indicate that 5% of Type A bulbs are defective, while only 1% 
    of Type B bulbs are defective. Now, a randomly selected bulb is tested and found to be defective. What is the probability 
    that this defective bulb is of Type A?

Solution using Bayes's Theorem:
Let's define the events:
A: The bulb is Type A.
B: The bulb is defective.

Given:
\( P(A) = 0.10 \) (Prior probability of selecting Type A bulb)
\( P(B|A) = 0.05 \) (Probability of the bulb being defective given it is of Type A)
\( P(\neg A) = 0.90 \) (Prior probability of selecting Type B bulb, complement of A)
\( P(B|\neg A) = 0.01 \) (Probability of the bulb being defective given it is not of Type A, i.e., Type B)

We want to find:
\( P(A|B) \) (Probability of the bulb being of Type A given that it is defective)

Applying Bayes's Theorem:

\[ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} \]

We need to calculate \( P(B) \) using the law of total probability:

\[ P(B) = P(B|A) \times P(A) + P(B|\neg A) \times P(\neg A) \]
\[ P(B) = 0.05 \times 0.10 + 0.01 \times 0.90 \]
\[ P(B) = 0.005 + 0.009 \]
\[ P(B) = 0.014 \]

Now we can find \( P(A|B) \):

\[ P(A|B) = \frac{0.05 \times 0.10}{0.014} \]
\[ P(A|B) = \frac{0.005}{0.014} \]
\[ P(A|B) ≈ 0.357 \]

So, the probability that the defective bulb is of Type A is approximately 0.357 or 35.7%.
```

```python
Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.
Ans:
    To calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation
    of 5, we use the formula for the z-score:

\[ \text{Z-score} = \frac{\text{Data point} - \text{Population Mean}}{\text{Population Standard Deviation}} \]

Given:
Data point (X) = 75
Population Mean (μ) = 70
Population Standard Deviation (σ) = 5

Now, let's calculate the z-score:

\[ \text{Z-score} = \frac{75 - 70}{5} = \frac{5}{5} = 1 \]

Interpretation:
The calculated z-score is 1. This means that the data point 75 is one standard deviation above the population mean of 70. The
positive sign of the z-score indicates that the data point is higher than the mean.

The z-score helps in understanding how far a particular data point is from the mean in terms of standard deviations. In this
case, a z-score of 1 tells us that the data point 75 is relatively close to the mean, just one standard deviation above it. If
the z-score were negative, it would mean the data point is below the mean. A z-score of 0 would indicate that the data point is 
at the mean. Z-scores are useful for comparing data points from different distributions and understanding their relative posit-
ions within their respective datasets.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.
Ans:
    To conduct a hypothesis test to determine if the weight loss drug is significantly effective, we will use a t-test since the
    sample size is small (n = 50) and the population standard deviation is unknown. We will set up the null and alternative hy-
    potheses and use the t-test to calculate the t-statistic and p-value.

Null Hypothesis (H0): The new weight loss drug has no significant effect, and the mean weight loss in the population is equal to
    or less than 0 pounds. μ ≤ 0.

Alternative Hypothesis (Ha): The new weight loss drug is significantly effective, and the mean weight loss in the population is
    greater than 0 pounds. μ > 0.

Next, we calculate the t-statistic using the formula:

\[ t = \frac{\text{Sample Mean} - \text{Population Mean}}{\frac{\text{Sample Standard Deviation}}{\sqrt{\text{Sample Size}}}} \]


Given:
Sample Mean (X̄) = 6 pounds
Sample Standard Deviation (s) = 2.5 pounds
Sample Size (n) = 50
Population Mean (μ) = 0 pounds (since the null hypothesis assumes no effect)

\[ t = \frac{6 - 0}{\frac{2.5}{\sqrt{50}}} = \frac{6}{0.354} \approx 16.95 \]

Now, we need to find the critical t-value for a one-tailed test with a 95% confidence level and 49 degrees of freedom (n - 1 =
    50 - 1 = 49). From the t-table or using a statistical software, the critical t-value is approximately 1.677 (for a one-tai-
                                                led test at α = 0.05).

Since the calculated t-statistic (16.95) is greater than the critical t-value (1.677), we reject the null hypothesis.

Conclusion:
At a 95% confidence level, we have enough evidence to conclude that the new weight loss drug is significantly effective, as the 
mean weight loss in the population is greater than 0 pounds.
```

```python
Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.
Ans:
    To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we use the 
    formula for the confidence interval of a proportion:

\[ \text{Confidence Interval} = \text{Sample Proportion} \pm (\text{Critical Value} \times \sqrt{\frac{\text{Sample Proportion}
                        \times (1 - \text{Sample Proportion})}{\text{Sample Size}}}) \]

Given:
Sample Proportion (p̂) = 0.65 (65% reported being satisfied)
Sample Size (n) = 500

The critical value for a 95% confidence level for a two-tailed test can be found using a standard normal distribution table or 
calculator, which is approximately 1.96.

Now, let's calculate the confidence interval:

\[ \text{Confidence Interval} = 0.65 \pm (1.96 \times \sqrt{\frac{0.65 \times (1 - 0.65)}{500}}) \]
\[ \text{Confidence Interval} = 0.65 \pm (1.96 \times \sqrt{\frac{0.65 \times 0.35}{500}}) \]
\[ \text{Confidence Interval} = 0.65 \pm (1.96 \times \sqrt{\frac{0.2275}{500}}) \]
\[ \text{Confidence Interval} = 0.65 \pm (1.96 \times 0.0201) \]
\[ \text{Confidence Interval} = 0.65 \pm 0.0393 \]

So, the 95% confidence interval for the true proportion of people who are satisfied with their job is approximately (0.6107, 0.
    6893). This means we are 95% confident that the true proportion of satisfied people lies within this range based on the sa-
mple data.
```

```python
Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
significant difference in student performance using a t-test with a significance level of 0.01.
Ans:
    To conduct a hypothesis test to determine if there is a significant difference in student performance between the two teach-
    ing methods, we will use a two-sample t-test. The significance level (α) is given as 0.01, which corresponds to a 99% conf-
    idence level. We will set up the null and alternative hypotheses and use the t-test to calculate the t-statistic and p-value
    .

Null Hypothesis (H0): There is no significant difference in student performance between the two teaching methods. The populat-
    ion means of sample A (μA) and sample B (μB) are equal. μA - μB = 0.

Alternative Hypothesis (Ha): There is a significant difference in student performance between the two teaching methods. The pop-
    ulation means of sample A and sample B are not equal. μA - μB ≠ 0.

Next, we calculate the pooled standard error and the t-statistic using the formula:

\[ \text{Pooled Standard Error} = \sqrt{\frac{\text{Sample A Standard Deviation}^2}{\text{Sample A Size}} + \frac{\text{Sample B
                                    Standard Deviation}^2}{\text{Sample B Size}}} \]

\[ \text{t-statistic} = \frac{\text{Sample A Mean} - \text{Sample B Mean}}{\text{Pooled Standard Error}} \]

Given:
Sample A Mean (X̄A) = 85
Sample A Standard Deviation (σA) = 6
Sample A Size (nA) = (not provided)
Sample B Mean (X̄B) = 82
Sample B Standard Deviation (σB) = 5
Sample B Size (nB) = (not provided)

Since the sample sizes (nA and nB) are not provided, we cannot calculate the pooled standard error and the t-statistic. The sam-
ple sizes are essential for conducting the t-test.

In a t-test, the sample size plays a crucial role in determining the variability of the sample mean and, consequently, the sign-
ificance of the results. Larger sample sizes generally lead to more precise estimates and can increase the power of the hypothe-
sis test to detect significant differences.

Without the sample sizes, we cannot proceed with the hypothesis test. We need the complete information on both sample sizes to 
calculate the pooled standard error, t-statistic, and p-value.
```
