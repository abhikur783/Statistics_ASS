```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.
Ans:
    
Probability Mass Function (PMF) and Probability Density Function (PDF) are mathematical concepts used in probability and 
statistics to describe the probability distribution of a random variable.

Probability Mass Function (PMF):
The Probability Mass Function (PMF) is used for discrete random variables. It gives the probability of each possible outcome in 
the sample space. In other words, it provides the probability that a random variable takes on a specific value.
Mathematically, for a discrete random variable X, the PMF is defined as:
PMF(X = x) = P(X = x)

Where:

PMF(X = x) is the probability of the random variable X taking on the value x.
P(X = x) is the probability of the event X = x occurring.
The PMF satisfies two properties:

For any value x in the sample space, 0 ≤ PMF(X = x) ≤ 1.
The sum of probabilities for all possible values in the sample space is equal to 1.
Example:
Consider rolling a fair six-sided die. The sample space for this random experiment is {1, 2, 3, 4, 5, 6}. The PMF for this 
experiment would be:

PMF(X = 1) = 1/6
PMF(X = 2) = 1/6
PMF(X = 3) = 1/6
PMF(X = 4) = 1/6
PMF(X = 5) = 1/6
PMF(X = 6) = 1/6

Probability Density Function (PDF):
The Probability Density Function (PDF) is used for continuous random variables. It gives the relative likelihood of the random
variable falling within a particular range of values. Unlike the PMF, which provides exact probabilities for specific values, 
the PDF provides probabilities as densities over intervals.
Mathematically, for a continuous random variable X, the PDF is denoted as f(x), and it satisfies the following properties:

f(x) ≥ 0 for all x.
The area under the PDF curve over its entire range is equal to 1.
The probability that X falls within a specific interval [a, b] is given by the integral of the PDF over that interval:
P(a ≤ X ≤ b) = ∫[a to b] f(x) dx

Example:
Consider the height of adult males in a population, which is a continuous random variable. The PDF for this variable might be a
bell-shaped curve, such as the normal distribution. Let's say the PDF for this population's height follows a standard normal 
distribution, which is given by:

f(x) = (1/√(2π)) * e^(-(x^2)/2)

In this case, the probability of a randomly chosen adult male having a height between 170 cm and 180 cm would be:

P(170 ≤ X ≤ 180) = ∫[170 to 180] f(x) dx

Note: To compute the actual probability within the interval, you need to evaluate the integral of the PDF between the given 
    limits.
```

```python
Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
Ans:
    
The Cumulative Density Function (CDF) is a fundamental concept used in probability and statistics to describe the cumulative 
probability distribution of a random variable. It provides the probability that a random variable takes on a value less than or 
equal to a given value.

Mathematically, for a random variable X, the CDF is denoted as F(x) and is defined as follows:

F(x) = P(X ≤ x)

Where:

F(x) is the cumulative probability that the random variable X takes on a value less than or equal to x.
P(X ≤ x) is the probability that the event X ≤ x occurs.
Properties of the CDF:

F(x) is a non-decreasing function. As x increases, F(x) does not decrease; it either stays constant or increases.
The limit of F(x) as x approaches negative infinity is 0 (F(-∞) = 0), and the limit as x approaches positive infinity is 1
(F(∞) = 1).
The CDF is right-continuous. This means that for any value x, the limit of F(t) as t approaches x from the right is equal to 
F(x).
Example:
Let's consider a simple example with a discrete random variable X representing the outcome of rolling a fair six-sided die. The
sample space for this random experiment is {1, 2, 3, 4, 5, 6}. The CDF for this experiment would be:

F(x) = P(X ≤ x)

For each value of x, we can calculate the cumulative probability as follows:

F(1) = P(X ≤ 1) = 1/6
F(2) = P(X ≤ 2) = 2/6 = 1/3
F(3) = P(X ≤ 3) = 3/6 = 1/2
F(4) = P(X ≤ 4) = 4/6 = 2/3
F(5) = P(X ≤ 5) = 5/6
F(6) = P(X ≤ 6) = 6/6 = 1

Now, if we plot these values on a graph, we would get a step function that increases at each value of x, reflecting the cumulat-
ive probabilities.

Why CDF is used?
The CDF is essential for several reasons:

Probability calculations: The CDF allows us to find probabilities for a random variable being less than or equal to a specific 
    value. This is useful for computing probabilities over intervals or finding the likelihood of certain events occurring.

Finding percentiles: The CDF can be used to determine percentiles, such as the median or quartiles, which are critical measures 
    in statistics.

Comparing distributions: The CDF allows for easy visual and analytical comparisons between different probability distributions.

Generating random numbers: The inverse of the CDF, known as the Quantile Function, allows us to generate random numbers followi-
    ng a particular probability distribution.

In summary, the Cumulative Density Function is a valuable tool in probability and statistics that provides a comprehensive unde-
rstanding of the probabilities associated with a random variable and enables various important calculations and analyses.
```

```python
Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.
Ans:
    
The normal distribution, also known as the Gaussian distribution, is one of the most widely used probability distributions in 
statistics. It is often used as a model for various real-world situations due to its many desirable properties. Some examples
of situations where the normal distribution might be used as a model include:

Heights and weights of a population: The normal distribution is commonly used to model the heights and weights of individuals 
    in a population. While not perfect, it provides a good approximation for these continuous variables.

IQ scores: IQ scores in a large population tend to follow a normal distribution, with the majority of scores clustered around 
    the mean.

Measurement errors: In many scientific experiments and measurements, errors are often modeled using a normal distribution, 
    assuming they are random and symmetrically distributed.

Natural phenomena: Many natural phenomena, such as the distribution of errors in physical measurements, the distribution of
    reaction times, and the distribution of certain biological characteristics, can often be approximated by a normal distribu-
    tion.

Financial markets: In finance, asset returns are often assumed to follow a normal distribution in many theoretical models, such
    as the famous Black-Scholes option pricing model.

The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ). The mean represents the
    center or the location of the distribution, while the standard deviation controls the spread or dispersion of the distribut-
    ion. Here's how these parameters relate to the shape of the normal distribution:

Mean (μ):
The mean determines the center of the normal distribution. When the mean is increased or decreased, the entire distribution 
shifts left or right along the x-axis, respectively. If μ = 0, the distribution is centered at the origin (symmetric about the
                                                                                                           y-axis).

Standard Deviation (σ):
The standard deviation controls the spread of the normal distribution. A larger standard deviation leads to a wider and flatter
curve, while a smaller standard deviation results in a narrower and taller curve. When σ = 0, the distribution collapses to a 
single point (degenerate case).

Generally, when both μ and σ are known, the normal distribution is fully specified. However, if only the mean μ is known and the
standard deviation σ is unknown, the normal distribution can still be used with σ being a parameter that can be estimated from 
data.

It's important to note that many real-world situations only approximately follow a normal distribution. In practice, the normal 
distribution is often used as an approximation when dealing with real data, especially when the sample size is large enough due
to the Central Limit Theorem.
```

```python
Q7: Explain uniform Distribution with an example. in sort answer
Ans:
    
The uniform distribution is a continuous probability distribution where all values in a given interval have an equal probability
of occurring. It is characterized by a constant probability density function (PDF) within the interval and zero probability out-
side that interval.

Example:
Rolling a fair six-sided die is an example of a discrete uniform distribution. The die has six faces, and each face has an equal
probability of 1/6 of landing face-up after a roll. This means that each outcome (1, 2, 3, 4, 5, or 6) has an equal chance of 
occurring, making it a uniform distribution.

In a continuous uniform distribution, a simple example could be selecting a random number between 0 and 1. Any value within this
interval has an equal chance of being selected, and the probability density function remains constant (equal to 1) within this 
range. Outside this interval, the probability is 0, as no values can be selected beyond the given range.
```

```python
Q8: What is the z score? State the importance of the z score. in sort form answer
Ans:
    
The z-score, also known as the standard score, is a statistical measure that quantifies how many standard deviations a data poi-
nt is from the mean of a distribution. It is calculated using the formula:

z = (X - μ) / σ

Where:

X is the individual data point or observation.
μ is the mean of the distribution.
σ is the standard deviation of the distribution.
Importance of the z-score:

Standardization: The z-score standardizes data, allowing for easy comparison of values from different distributions with varying
    scales and units. It transforms data to a common scale, centered around the mean and expressed in terms of standard deviat-
    ions.

Identifying Outliers: Z-scores help identify outliers or extreme values in a dataset. Observations with z-scores far from zero 
    are considered unusual or anomalous in comparison to the rest of the data.

Probability Calculation: The z-score is used to calculate probabilities and find percentiles in a standard normal distribution,
    which has a mean of 0 and a standard deviation of 1.

Hypothesis Testing: Z-scores are essential in hypothesis testing, allowing researchers to determine whether a sample mean signif
    -icantly differs from a known population mean.

Data Analysis and Inference: Z-scores aid in making data-driven decisions and drawing inferences about a population based on 
    sample statistics.

In summary, the z-score is a valuable tool in statistics that standardizes data, enables comparisons, and plays a significant
role in various statistical analyses and probability calculations.
```

```python
Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem. in sort form answer
Ans:
    
The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that the sampling distribution of the sample
mean from a large number of independent and identically distributed random samples will approximate a normal distribution,
regardless of the shape of the original population distribution.

Significance of the Central Limit Theorem:

Allows Approximation: The CLT is essential because it allows researchers to approximate the distribution of sample means as 
    normal, even when the underlying population distribution is not normal. This simplifies statistical analysis and inference.

Large Sample Size: The CLT indicates that as the sample size increases, the sampling distribution of the sample mean becomes 
    more and more normally distributed, making the normal approximation increasingly accurate.

Foundation for Inference: The CLT forms the basis for many statistical inference techniques, such as hypothesis testing and 
    confidence intervals, which rely on the assumption of normality for sample means.

Real-world Applications: The CLT is widely used in various fields, allowing researchers to make statistical inferences and draw
    conclusions from sample data, even when the population distribution is unknown or not normally distributed.

Widely Applicable: The CLT applies to a broad range of distributions, making it a powerful and versatile tool in statistics.

In summary, the Central Limit Theorem is a critical concept in statistics that justifies the use of normal distribution approx-
imations for sample means, enabling researchers to draw reliable conclusions from data and perform various statistical analyses
effectively.
```

```python
Q10: State the assumptions of the Central Limit Theorem. in sort answer
Ans:
    
The assumptions of the Central Limit Theorem (CLT) are as follows:

Random Sampling: The samples must be selected randomly from the population of interest. Each observation in the sample should be
    independent of one another.

Sample Size: The sample size should be sufficiently large. While there is no strict rule for what constitutes a "large" sample 
    size, as a general guideline, a sample size of 30 or more is often considered adequate for the CLT to hold, though it may 
    vary depending on the original distribution's characteristics.

Identically Distributed: The samples should be drawn from the same population, and each sample should have the same underlying 
    probability distribution.

Under these assumptions, the Central Limit Theorem ensures that the sampling distribution of the sample mean will be approximat-
ely normally distributed, regardless of the shape of the original population distribution. This normal approximation is increas-
ingly accurate with larger sample sizes.
```
